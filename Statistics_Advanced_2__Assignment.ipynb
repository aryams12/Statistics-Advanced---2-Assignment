#Question 1: What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It helps determine whether there is enough evidence to support a specific claim or assumption.

In hypothesis testing, we:


1.   Form hypotheses
1.   Collect sample data
2.   Perform a statistical test
2.   Decide whether to reject or accept the hypothesis



#Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?


**1.   Null Hypothesis (H₀):**
States that there is no effect, no difference, or no change.
**2.   Alternative Hypothesis (H₁ or Hₐ):**
States that there is an effect, difference, or change.

**Example**

H₀: Average weight = 50 kg

H₁: Average weight ≠ 50 kg


#Question 3: Explain the significance level in hypothesis testing and its role.

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

1.  Common values: 0.05 (5%), 0.01
1.   If p-value ≤ α, reject H₀
2.   If p-value > α, fail to reject H₀

It controls the risk of making a Type I error.


#Question 4: What are Type I and Type II errors? Give examples.

**Type I Error**

1.   Meaning: Rejecting a true null hypothesis

Type II Error

#Question 5: Difference between Z-test and T-test
**Z-test**

1.   Population std dev known
1.   Large sample (n ≥ 30)
2.   Uses normal distribution

**T-test**


1.   Population std dev unknown
1.   Small sample (n < 30)
2.   Uses t-distribution

**Usage:**

Use Z-test for large samples

Use T-test for small samples

```python
#Question 7: Hypothesis Testing using Z-statistics

#Hypotheses

#H0: μ = 50

#H1: μ ≠ 50

α = 0.05

import numpy as np
import scipy.stats as stats

sample_data = np.array([
49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9
])

mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

z = (mean - 50) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p_value)
```

```python
#Question 8: Normal Distribution & 95% Confidence Interval

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(100, 15, 1000)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = stats.norm.ppf(0.975)
margin = z * (std / np.sqrt(n))

print("95% Confidence Interval:", (mean - margin, mean + margin))

plt.hist(data, bins=30)
plt.title("Normal Distribution Data")
plt.show()

#95% Confidence Interval: (99.1, 100.8)
```

```python
#Question 9: Z-score Calculation & Visualization

#Explanation

#A Z-score tells how many standard deviations a value is away from the mean.

#Z = 0 → at the mean

#Z = +1 → 1 SD above mean

#Z = −1 → 1 SD below mean

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(50, 5, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores")
plt.show()

print("Sample Z-scores:", z_scores[:5])
```
